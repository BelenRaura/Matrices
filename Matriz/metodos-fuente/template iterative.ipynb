```python
A = [[10, 2],
    [2, 3]]

# Vector de términos independientes
b = [16, 11]
n=len(A)
r2 = gauss_jacobi(A=A,b=b,x0= [1]*n, tol= 1e-1, max_iter= 100)
# Convertir la trayectoria a un array de NumPy
trayectoria_jacobi = np.array(r2[1])

# Graficar la trayectoria hasta que converge
plt.figure(figsize=(8, 6))
plt.plot(trayectoria_jacobi[:, 0], trayectoria_jacobi[:, 1], marker='o', linestyle='-', label='Trayectoria Jacobi')
plt.scatter(trayectoria_jacobi[-1, 0], trayectoria_jacobi[-1, 1], color='red', label='Solución final')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Trayectoria del método de Jacobi en el espacio de soluciones hasta converger')
plt.legend()
plt.grid(True)
plt.show()
```
